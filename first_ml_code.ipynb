```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading dataset
df = pd.read_csv("Coffe_Sales.csv")
```

```python
#Cleaning data 
#Checking for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
#Checking for missing values
print("\nMissing values before cleaning :", df.isna().sum())
```

```python
# Display the first 5 rows to see what the data looks like
print("\n--- First 5 rows of the dataset ---")
print(df.head())

# Get a concise summary of the DataFrame
# This shows the data types of each column and the number of non-null values
print("\n--- DataFrame Information ---")
df.info()

# Get descriptive statistics for numerical columns
# This provides insights into count, mean, standard deviation, etc.
print("\n--- Descriptive Statistics for Numerical Columns ---")
print(df.describe())
```

```python
# Check unique values and their frequencies for a categorical column, e.g., 'coffee_name'
print("\n--- Value Counts for 'coffee_name' ---")
print(df['coffee_name'].value_counts())

# You can do this for any other categorical column as well
print("\n--- Value Counts for 'Time_of_Day' ---")
print(df['Time_of_Day'].value_counts())
```

```python
coffee_counts = df['coffee_name'].value_counts()
print("\nMost sold coffees",coffee_counts)
time_count = df['Time_of_Day'].value_counts()
print("\nTime of the day coffee was sold",time_count)
day_count = df['Weekday'].value_counts()
print("\nWeekday when most people enjoyed coffee",day_count)
month_count = df['Month_name'].value_counts()
print("\nMonth when people drank coffee most", month_count)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=coffee_counts.index, y=coffee_counts.values, palette='bright')
plt.title('Sales Count of Different Coffee Names', fontsize=16)
plt.xlabel('Coffee Name', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=time_count.index, y=time_count.values, palette='deep')
plt.title('Time of the day when people enjoyed the coffee most', fontsize=18)
plt.xlabel('Time of the day', fontsize=14)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=day_count.index, y=day_count.values, palette='dark')
plt.title('Weekday when people loved the coffee most', fontsize=18)
plt.xlabel('Weekday', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=month_count.index, y=month_count.values, palette='Accent')
plt.title('Month when coffee sold most', fontsize=15)
plt.xlabel('Month name', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

```python
date_count = df['Date'].value_counts()
top_10_dates = date_count.head(10)
print("Top 10 days in the year when coffee was sold most\n", top_10_dates)
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x=top_10_dates.index, y=top_10_dates.values, palette='icefire')
plt.title('Dates when people drank most coffee', fontsize=18)
plt.xlabel('Dates', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```
